```python
# finding length of a string without using inbuilt functions
def findlength(st):
    """ this function helps to find the length of a string """
    count = 0
    for i in st:
        count+=1
    print(f"Length of a string is {count}")
st = input("Enter String:")
findlength(st)
```

```python
#finding index of each element inside a list
def findIndex(li):
    """ this function helps to find the index of each element inside a list """
    for i in range(0,len(li)):
        print(f"Index of {li[i]} is {i}")
li = ["hello","ineuron",1,2,3,2.45,0,9]
findIndex(li)
```

```python
#find multiplication of all elements present inside a list
def findMultiply(l):
    """ this function helps to find the multiplication of a elements inside a list """
    mul = 1
    for i in l:
        if type(i) == int or type(i) == float:
            mul*=i
    print(f"Multiplication of list elements is {mul}")
l = [3.5, 6.56, 4,5,"sudh" , "ineuron" , 'fsda bootcamp 2.0']
findMultiply(l)
```

```python
#read mails from inbox
import smtplib
import time
import imaplib
import email
import traceback
ORG_EMAIL = "@gmail.com" 
FROM_EMAIL = 'manishguru72@gmail.com'
FROM_PWD = 'rruyb1234vistry'
SMTP_SERVER = "imap.gmail.com" 
SMTP_PORT = 993
imaplib._MAXLINE = 10000000000

def read_email_from_gmail():
    """ this function reads the inbox mails from specified email """
    try:
        mail = imaplib.IMAP4_SSL(SMTP_SERVER)
        mail.login(FROM_EMAIL,FROM_PWD)
        mail.select('inbox')

        data = mail.search(None, 'ALL')
        mail_ids = data[1]
        id_list = mail_ids[0].split()   
        first_email_id = int(id_list[0])
        latest_email_id = int(id_list[-1])

        for i in range(latest_email_id,first_email_id, -1):
            data = mail.fetch(str(i), '(RFC822)' )
            for response_part in data:
                arr = response_part[0]
                if isinstance(arr, tuple):
                    msg = email.message_from_string(str(arr[1],'utf-8'))
                    email_subject = msg['subject']
                    email_from = msg['from']
                    print('From : ' + email_from + '\n')
                    print('Subject : ' + email_subject + '\n')

    except Exception as e:
        traceback.print_exc() 
        print(str(e))

read_email_from_gmail()
```

```python
#send an email
import smtplib, ssl
def sendemail():
  """ this function sends an email to specified user """
  port = 465  # For SSL
  smtp_server = "smtp.gmail.com"
  sender_email = "manishguru72@gmail.com"  # Enter your address
  receiver_email = "manishvankadhara72@gmail.com"  # Enter receiver address
  password = 'rruyb1234vistry'
  message = """this is my message from python code """

  context = ssl.create_default_context()
  with smtplib.SMTP_SSL(smtp_server, port, context=context) as server:
      server.login(sender_email, password)
      server.sendmail(sender_email, receiver_email, message)

sendemail()
```
